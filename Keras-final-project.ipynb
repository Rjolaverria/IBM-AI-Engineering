<h1 align=center><font size = 5>A - Build a baseline model</font></h1>

Load and clean data

```python
# Import sklearn, pandas and numpy for manipulating data
import pandas as pd
import numpy as np
from sklearn import model_selection
```

```python
# Dwonload csv data to pandas dataframe
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
# Split predictors and target
predictors = concrete_data[concrete_data.columns[concrete_data.columns != 'Strength']]
target = concrete_data['Strength']

input_count = predictors.shape[1]
```

```python
# Import Keras
import keras
from keras.models import Sequential
from keras.layers import Dense
```

Train/Test model 50 times. Calculate the Mean and Standard from the list of Mean Squared Errors.

```python
# List of Mean Squared Errors
mse_list = []

# Train and test 50 times
for i in range(50):

    # Train/Test split
    X_train, X_test, y_train, y_test = model_selection.train_test_split(
        predictors, target, test_size=0.3
    )

    # Create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(input_count,)))
    model.add(Dense(1))

    # Compile
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Fit training data
    model.fit(X_train, y_train, epochs=50, verbose=2)

    # Test model
    from sklearn.metrics import mean_squared_error
    prediction = model.predict(X_test, verbose=0)

    # Add MSE to list
    mse_list.append(mean_squared_error(y_test,prediction))
```

```python
# Convert list to numpy array
mse_list = np.array(mse_list)

# Calculate the mean and standard deviation.
mean = mse_list.mean()
stdev = mse_list.std()

print(f"Mean: {mean:.2f} , Standard deviation: {stdev:.2f}")
```

<h1 align=center><font size = 5>B - Normalize the data </font></h1>

Repeat with Normalized data

```python
normalized = (predictors - predictors.mean()) / predictors.std()

# List of Mean Squared Errors
mse_list = []

# Train and test 50 times
for i in range(50):

    # Train/Test split
    X_train, X_test, y_train, y_test = model_selection.train_test_split(
        normalized, target, test_size=0.3
    )

    # Create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(input_count,)))
    model.add(Dense(1))

    # Compile
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Fit training data
    model.fit(X_train, y_train, epochs=50, verbose=2)

    # Test model
    prediction = model.predict(X_test, verbose=0)

    # Add MSE to list
    mse_list.append(mean_squared_error(y_test,prediction))
```

```python
mse_list = np.array(mse_list)
mean = mse_list.mean()
stdev = mse_list.std()

print(f"Mean: {mean:.2f} , Standard deviation: {stdev:.2f}")
```

** Normalizing the data increases the mean but the Standard deviation decreased.

<h1 align=center><font size = 5>C - Increase number of Epochs </font></h1>

Repeat with 100 epochs

```python
# List of Mean Squared Errors
mse_list = []

# Train and test 50 times
for i in range(50):

    # Train/Test split
    X_train, X_test, y_train, y_test = model_selection.train_test_split(
        normalized, target, test_size=0.3
    )

    # Create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(input_count,)))
    model.add(Dense(1))

    # Compile
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Fit training data
    model.fit(X_train, y_train, epochs=100, verbose=2)

    # Test model
    prediction = model.predict(X_test, verbose=0)

    # Add MSE to list
    mse_list.append(mean_squared_error(y_test,prediction))
```

```python
mse_list = np.array(mse_list)
mean = mse_list.mean()
stdev = mse_list.std()

print(f"Mean: {mean:.2f} , Standard deviation: {stdev:.2f}")
```

** With more Epochs the mean decreases by about half.

<h1 align=center><font size = 5>D - Increase the number of hidden layers </font></h1>

Three hidden layers, each of 10 nodes and ReLU activation.

```python
# List of Mean Squared Errors
mse_list = []

# Train and test 50 times
for i in range(50):

    # Train/Test split
    X_train, X_test, y_train, y_test = model_selection.train_test_split(
        normalized, target, test_size=0.3
    )

    # Create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(input_count,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))

    # Compile
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Fit training data
    model.fit(X_train, y_train, epochs=100, verbose=2)

    # Test model
    prediction = model.predict(X_test, verbose=0)

    # Add MSE to list
    mse_list.append(mean_squared_error(y_test,prediction))
```

```python
mse_list = np.array(mse_list)
mean = mse_list.mean()
stdev = mse_list.std()

print(f"Mean: {mean:.2f} , Standard deviation: {stdev:.2f}")
```

** Adding more hidden layers decreased the Mean by almost half again.
